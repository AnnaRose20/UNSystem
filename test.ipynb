```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def get_icj_direct_pdfs(start_case, end_case):
    base_case_url = "https://www.icj-cij.org"
    all_pdf_data = []

    print(f"Extracting direct PDF links for cases {start_case} to {end_case}...")

    for case_num in range(start_case, end_case + 1):
        case_url = f"{base_case_url}{case_num}"
        try:
            response = requests.get(case_url, timeout=15)
            if response.status_code != 200:
                continue
            
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Find every link ending in .pdf
            pdf_links = soup.find_all('a', href=True)
            
            for link in pdf_links:
                url = link['href']
                if url.endswith('.pdf'):
                    # Ensure the URL is absolute
                    full_pdf_url = url if url.startswith('http') else f"https://www.icj-cij.org{url}"
                    
                    # Extract the filename for metadata
                    filename = url.split('/')[-1]
                    
                    all_pdf_data.append({
                        "Case Number": case_num,
                        "Document Name": link.text.strip() or filename,
                        "Filename": filename,
                        "Direct PDF URL": full_pdf_url
                    })
            
            print(f"Finished Case {case_num}: Found {len([d for d in all_pdf_data if d['Case Number'] == case_num])} PDFs")
            time.sleep(1) # Delay to respect the server

        except Exception as e:
            print(f"Error on Case {case_num}: {e}")

    return pd.DataFrame(all_pdf_data)

# Case 162 (2023) to current Case 201 (2025/2026)

df_icj_pdfs = get_icj_direct_pdfs(162, 201)

# Save to Excel
output_file = "ICJ_Direct_PDF_Links_2016_2026.xlsx"
df_icj_pdfs.to_excel(output_file, index=False)
print(f"\nSaved {len(df_icj_pdfs)} direct PDF links to {output_file}")
```

```python
import pandas as pd

# Configuration for 2016-2026 (Sessions 31 to 61)
# Session 61 is current (Feb-Mar 2026)
config = {
    "sessions": range(31, 62),
    "ga_sessions": range(71, 81), # General Assembly sessions for 2016-2026
    "max_res": 45,      # High estimate for resolutions per session
    "max_dec": 5,       # Decisions per session
    "max_prst": 3,      # President's statements per session
    "max_sr": 60,       # Meetings (Summary Records) per session
    "max_l": 50         # Draft resolutions (L documents) per session
}

def generate_hrc_links():
    base_url = "https://undocs.org/en"
    links = []

    for sess in config["sessions"]:
        # 1. Resolutions, Decisions, and PRSTs
        for r in range(1, config["max_res"] + 1):
            links.append({"Type": "Resolution", "Symbol": f"A/HRC/RES/{sess}/{r}"})
        for d in range(1, config["max_dec"] + 1):
            links.append({"Type": "Decision", "Symbol": f"A/HRC/DEC/{sess}/{d}"})
        for p in range(1, config["max_prst"] + 1):
            links.append({"Type": "President's Statement", "Symbol": f"A/HRC/PRST/{sess}/{p}"})

        # 2. Sessional Reports (A/HRC/[Sess]/2 is standard for the report)
        links.append({"Type": "Sessional Report", "Symbol": f"A/HRC/{sess}/2"})

        # 3. Summary Records (SR)
        # Note: SRs are only taken for certain meetings (action/adoption) since sess 48
        for sr in range(1, config["max_sr"] + 1):
            links.append({"Type": "Summary Record (SR)", "Symbol": f"A/HRC/{sess}/SR.{sr}"})

        # 4. Draft Resolutions (L documents)
        for l_doc in range(1, config["max_l"] + 1):
            links.append({"Type": "Draft Resolution (L)", "Symbol": f"A/HRC/{sess}/L.{l_doc}"})

    # 5. Annual Reports to GA (Supplement 53)
    for ga_sess in config["ga_sessions"]:
        links.append({"Type": "Annual Report to GA", "Symbol": f"A/{ga_sess}/53"})

    # Build final URL list
    df = pd.DataFrame(links)
    df["URL"] = df["Symbol"].apply(lambda s: f"{base_url}/{s}")
    return df

# Generate and save
df_hrc = generate_hrc_links()
df_hrc.to_excel("HRC_Official_Documents_2016_2026.xlsx", index=False)
print(f"Generated {len(df_hrc)} links for HRC documents.")
```

```python
import datetime
import pandas as pd

def generate_un_habitat_excel(years_back=10):
    current_year = datetime.datetime.now().year
    start_year = current_year - years_back
    rows = []

    for year in range(start_year, current_year + 1):
        # 1. Executive Board Documents (Modern series: HSP/EB/YYYY/1-20)
        # These cover the Strategic Plan, Budget, and Annual Reports
        for i in range(1, 21):
            symbol = f"HSP/EB/{year}/{i}"
            rows.append({
                "Year": year, 
                "Type": "Executive Board Document", 
                "Symbol": symbol, 
                "Link": f"https://undocs.org/en/{symbol}"
            })

        # 2. Committee of Permanent Representatives (CPR)
        # Crucial for inter-sessional governance
        for i in range(1, 11):
            symbol = f"HSP/CPR/{year}/{i}"
            rows.append({
                "Year": year, 
                "Type": "CPR Working Document", 
                "Symbol": symbol, 
                "Link": f"https://undocs.org/en/{symbol}"
            })

        # 3. UN-Habitat Assembly (Every 4 years - e.g., 2019, 2023)
        # If the year matches an Assembly year, generate those links
        if year in [2019, 2023, 2027]:
            session_num = 1 if year == 2019 else (2 if year == 2023 else 3)
            for i in range(1, 15):
                symbol = f"HSP/HA.{session_num}/{i}"
                rows.append({
                    "Year": year, 
                    "Type": "Assembly Document", 
                    "Symbol": symbol, 
                    "Link": f"https://undocs.org/en/{symbol}"
                })

    # Create DataFrame and Save to Excel
    df = pd.DataFrame(rows)
    filename = "UN_Habitat_Official_Documents.xlsx"
    df.to_excel(filename, index=False)
    
    print(f"Excel file created: {filename}")
    return filename

# Run the script
generate_un_habitat_excel(10)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def get_fao_session_links(session_url):
    headers = {'User-Agent': 'Mozilla/5.0'}
    try:
        response = requests.get(session_url, headers=headers)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        
        links = []
        # FAO often lists documents in <a> tags with 'pdf' in the href
        for a in soup.find_all('a', href=True):
            if '.pdf' in a['href'].lower():
                full_link = a['href']
                if not full_link.startswith('http'):
                    full_link = f"https://www.fao.org{full_link}"
                
                links.append({
                    "Title": a.get_text(strip=True) or "Document",
                    "Link": full_link
                })
        return pd.DataFrame(links)
    except Exception as e:
        print(f"Error: {e}")
        return pd.DataFrame()

# Example: Get links for the 172nd Council Session
url = "https://www.fao.org"
df = get_fao_session_links(url)
df.to_excel("FAO_Session_Documents.xlsx", index=False)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def download_un_resolution(url, output_filename=None):
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    # Step 1: Load HTML page
    response = requests.get(url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    # Step 2: Extract embedded PDF link
    pdf_url = None

    iframe = soup.find("iframe")
    if iframe and iframe.get("src"):
        pdf_url = urljoin(url, iframe["src"])

    if not pdf_url:
        for a in soup.find_all("a", href=True):
            if ".pdf" in a["href"].lower():
                pdf_url = urljoin(url, a["href"])
                break

    if not pdf_url:
        print("❌ Could not find PDF link")
        return

    print("Found PDF:", pdf_url)

    # Step 3: Download PDF
    pdf_response = requests.get(pdf_url, headers=headers, stream=True)
    pdf_response.raise_for_status()

    # Ensure it's actually a PDF
    content_type = pdf_response.headers.get("Content-Type", "")
    if "pdf" not in content_type.lower():
        print("❌ The file is not a valid PDF")
        return

    # Step 4: Set filename
    if not output_filename:
        # Default filename from URL
        output_filename = pdf_url.split("/")[-1]

    if not output_filename.lower().endswith(".pdf"):
        output_filename += ".pdf"

    # Step 5: Save file
    with open(output_filename, "wb") as f:
        for chunk in pdf_response.iter_content(8192):
            f.write(chunk)

    print(f"✅ Saved as: {output_filename}")


# Example usage
download_un_resolution(
    "https://docs.un.org/en/A/RES/79/110",
    output_filename="A-RES-79-110.pdf"
)
```
